```python
# let's put the preparation above into a function to make it easier in the future

def read_dataframe(filename):
    if filename.endswith('.csv'):
        df = pd.read_csv(filename)

        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    elif filename.endswith('.parquet'):
        df = pd.read_parquet(filename)

    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    
    return df

```

```python
# let's try a different model, a kind of LR with regularisation
from sklearn.linear_model import Lasso

df_train = read_dataframe('mlops-zoomcamp/01-intro/data/green_tripdata_2021-01.parquet')
df_val = read_dataframe('mlops-zoomcamp/01-intro/data/green_tripdata_2021-02.parquet')
len(df_train), len(df_val)
```

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']

dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
# it's even worse with the val data, though this is our base model
```

```python
lr = Lasso()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
# even worse with the default params, let's see if we can adjust something
```

```python
lr = Lasso(0.001)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
# doesn't help much
```

```python
from sklearn.linear_model import Ridge

lr = Ridge(alpha=10)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
# pretty much the same thing... at least it's fast
```

```python
# Maybe we can actually be smart and combine the pickup and dropoff locations

df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID'].astype(str)
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID'].astype(str)
```

```python
categorical = ['PU_DO']
numerical = ['trip_distance']

dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)

target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
# much better
```
